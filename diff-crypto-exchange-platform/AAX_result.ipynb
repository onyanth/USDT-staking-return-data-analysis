```python
# Use Selenium to access the HTML
URL = "https://invest.aax.com/en-US/savings/"  

driver = webdriver.Chrome('./chromedriver' )
driver.get(URL)  # Get the HTML data of the webpage
subhtml = driver.page_source

'''
tag_day = driver.find_element_by_xpath('where is the %')
where is the xpath (address) => you find out the object and click the "..." at the left and choose copy > copy xpath
elelink_day = driver.find_element_by_xpath('where is the hyperlink')
.click() => click the hyperlink
time.sleep() => Go too fast in web scraping may lead to error
'''

all_data = []

time.sleep(2)
tag_360 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[2]')
print(tag_360.text)
all_data.insert(0,tag_360.text)

time.sleep(2)
eleLink_180 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[3]/div/div[6]')
eleLink_180.click()
tag_180 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[2]')
print(tag_180.text)
all_data.insert(0,tag_180.text)

time.sleep(2)
eleLink_90 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[3]/div/div[5]')
eleLink_90.click()
tag_90 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[2]')
print(tag_90.text)
all_data.insert(0,tag_90.text)

time.sleep(2)
eleLink_60 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[3]/div/div[4]')
eleLink_60.click()
tag_60 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[2]')
print(tag_60.text)
all_data.insert(0,tag_60.text)

time.sleep(2)
eleLink_30 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[3]/div/div[3]')
eleLink_30.click()
tag_30 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[2]')
print(tag_30.text)
all_data.insert(0,tag_30.text)

time.sleep(2)
eleLink_14 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[3]/div/div[2]')
eleLink_14.click()
tag_14 = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div[4]/div[2]')
print(tag_14.text)
all_data.insert(0,tag_14.text)

time.sleep(2)
eleLink_flex = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[1]/div/div/button[2]')
eleLink_flex.click()
tag_fle = driver.find_element_by_xpath('//*[@id="gatsby-focus-wrapper"]/div[2]/div[1]/div[3]/div[4]/div[1]/div/div[2]')
print(tag_fle.text)
all_data.insert(0,tag_fle.text)

time.sleep(2)
print(all_data)
```
