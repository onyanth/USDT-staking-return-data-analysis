```python
# Use Selenium to access the HTML
URL = "https://www.huobi.com/zh-hk/financial/earn/?type=steady"

driver = webdriver.Chrome('./chromedriver' )
driver.get(URL)  # Get the HTML data of the webpage
subhtml = driver.page_source

'''
tag_day = driver.find_element_by_xpath('where is the %')
where is the xpath (address) => you find out the object and click the "..." at the left and choose copy > copy xpath
elelink_day = driver.find_element_by_xpath('where is the hyperlink')
.click() => click the hyperlink
time.sleep() => Go too fast in web scraping may lead to error
'''

all_data = []

time.sleep(1)
tag_90 = driver.find_element_by_xpath('//*[@id="__layout"]/section/div[2]/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/table/tbody/tr[1]/td[2]/div/div/div')
print(tag_90.text)
all_data.insert(0,tag_90.text)

time.sleep(1)
eleLink_ch_30 = driver.find_element_by_xpath('//*[@id="__layout"]/section/div[2]/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/table/tbody/tr[1]/td[3]/div/div/div[3]/div/span')
eleLink_ch_30.click()

time.sleep(1)
tag_30 = driver.find_element_by_xpath('//*[@id="__layout"]/section/div[2]/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/table/tbody/tr[1]/td[2]/div/div/div')
print(tag_30.text)
all_data.insert(0,tag_30.text)

time.sleep(1)
eleLink_ch_14 = driver.find_element_by_xpath('//*[@id="__layout"]/section/div[2]/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/table/tbody/tr[1]/td[3]/div/div/div[2]/div/span')
eleLink_ch_14.click()

time.sleep(1)
tag_14 = driver.find_element_by_xpath('//*[@id="__layout"]/section/div[2]/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/table/tbody/tr[1]/td[2]/div/div/div')
print(tag_14.text)
all_data.insert(0,tag_14.text)

time.sleep(1)
eleLink_ch_flex = driver.find_element_by_xpath('//*[@id="__layout"]/section/div[2]/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/table/tbody/tr[1]/td[3]/div/div/div[1]/div/span')
eleLink_ch_flex.click()

time.sleep(1)
tag_flex = driver.find_element_by_xpath('//*[@id="__layout"]/section/div[2]/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/table/tbody/tr[1]/td[2]/div/div/div')
print(tag_flex.text)
all_data.insert(0,tag_flex.text)

print(all_data)

```

```python
import re
t_all_data = ['2.00%', '5.50%', '6.00%', '6.50%']
r_all_data = []
percentage = r'\d*.?\d*'  # take the number near the decimal point
for i in range(0,len(all_data)):
    data = re.findall(percentage,all_data[i])
    r_all_data.append(data[0])
    r_all_data[i] = float(r_all_data[i])    

print(r_all_data)
```
