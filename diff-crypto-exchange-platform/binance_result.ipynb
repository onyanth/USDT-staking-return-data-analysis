```python
# Use Selenium to access the HTML
URL = "https://www.binance.com/en/savings#lending-demandDeposits"
driver = webdriver.Chrome('./chromedriver')
driver.get(URL)  # Get the HTML data of the webpage
subhtml = driver.page_source

'''
tag_day = driver.find_element_by_xpath('where is the %')
where is the xpath (address) => you find out the object and click the "..." at the left and choose copy > copy xpath
elelink_day = driver.find_element_by_xpath('where is the hyperlink')
.click() => click the hyperlink
time.sleep() => Go too fast in web scraping may lead to error
'''

all_data = []
time.sleep(1)
eleLink_cookie = driver.find_element_by_xpath('//*[@id="onetrust-accept-btn-handler"]')
eleLink_cookie.click()

time.sleep(1)
txtbox = driver.find_element_by_xpath('//*[@id="savings-search-coin"]')
txtbox.send_keys('USDT')

time.sleep(1)
eleLink_60 = driver.find_element_by_xpath('//*[@id="__APP"]/div/main/div/div[3]/div[5]/div[2]/div[1]/div[2]/div[2]/div[2]/div/button[2]')
eleLink_60.click()
tag_60 = driver.find_element_by_xpath('//*[@id="__APP"]/div/main/div/div[3]/div[5]/div[2]/div[1]/div[2]/div[1]/div[2]/div')
print(tag_60.text)
all_data.insert(0,tag_60.text)

time.sleep(1)
eleLink_30 = driver.find_element_by_xpath('//*[@id="__APP"]/div/main/div/div[3]/div[5]/div[2]/div[1]/div[2]/div[2]/div[2]/div/button[1]')
eleLink_30.click()
tag_30 = driver.find_element_by_xpath('//*[@id="__APP"]/div/main/div/div[3]/div[5]/div[2]/div[1]/div[2]/div[1]/div[2]/div')
print(tag_30.text)
all_data.insert(0,tag_30.text)

print(all_data)
```

```python
import re
t_all_data = ['5\n%', '9\n%']
r_all_data = []
percentage = r'\d*'  #take out the number
for i in range(0,len(all_data)):
    data = re.findall(percentage,all_data[i])
    r_all_data.append(data[0])
    r_all_data[i] = float(r_all_data[i])
    
print(r_all_data)
```
