```python
# Use Selenium to access the HTML
URL = "https://crypto.com/us/earn"

driver = webdriver.Chrome('./chromedriver' )
driver.get(URL)  # Get the HTML data of the webpage
subhtml = driver.page_source

'''
tag_day = driver.find_element_by_xpath('where is the %')
where is the xpath (address) => you find out the object and click the "..." at the left and choose copy > copy xpath
elelink_day = driver.find_element_by_xpath('where is the hyperlink')
ele.send_keys => press the key in your keybroad
.click() => click the hyperlink
time.sleep() => Go too fast in web scraping may lead to error
'''

all_data = []

time.sleep(1)
eleLink_cookie = driver.find_element_by_xpath('//*[@id="onetrust-accept-btn-handler"]')
eleLink_cookie.click()

ele = driver.find_element_by_tag_name('body')
ele.send_keys(Keys.PAGE_DOWN)
ele.send_keys(Keys.PAGE_DOWN)

time.sleep(1)
eleLink_token = driver.find_element_by_xpath('//*[@id="token-select-inputbox"]')
eleLink_token.click()

time.sleep(1)
eleLink_token_select = driver.find_element_by_xpath('//*[@id="calculator"]/div/div[1]/div/div[2]/div[1]/div/div[2]/div[2]/div[4]/p')
eleLink_token_select.click()

ele.send_keys(Keys.PAGE_DOWN)
ele.send_keys(Keys.PAGE_DOWN)

time.sleep(1)
tag_90 = driver.find_element_by_xpath('//*[@id="reward-section"]/div/div[2]/div/div/div[4]/div/div[2]/p')
print(tag_90.text)
all_data.insert(0,tag_90.text)

ele.send_keys(Keys.PAGE_UP)
ele.send_keys(Keys.PAGE_UP)

time.sleep(1)
eleLink_ch_30_1 = driver.find_element_by_xpath('//*[@id="calculator"]/div/div[1]/div/div[2]/div[3]/div/div[2]/button')
eleLink_ch_30_1.click()
eleLink_ch_30_2 = driver.find_element_by_xpath('//*[@id="calculator"]/div/div[1]/div/div[2]/div[3]/div/div[2]/div/div[2]/p')
eleLink_ch_30_2.click()

ele.send_keys(Keys.PAGE_DOWN)
ele.send_keys(Keys.PAGE_DOWN)

time.sleep(1)
tag_30 = driver.find_element_by_xpath('//*[@id="reward-section"]/div/div[2]/div/div/div[4]/div/div[2]/p')
print(tag_30.text)
all_data.insert(0,tag_30.text)

ele.send_keys(Keys.PAGE_UP)
ele.send_keys(Keys.PAGE_UP)

time.sleep(1)
eleLink_ch_fl_1 = driver.find_element_by_xpath('//*[@id="calculator"]/div/div[1]/div/div[2]/div[3]/div/div[2]/button')
eleLink_ch_fl_1.click()
eleLink_ch_fl_2 = driver.find_element_by_xpath('//*[@id="calculator"]/div/div[1]/div/div[2]/div[3]/div/div[2]/div/div[1]/p')
eleLink_ch_fl_2.click()

ele.send_keys(Keys.PAGE_DOWN)
ele.send_keys(Keys.PAGE_DOWN)

time.sleep(1)
tag_fl = driver.find_element_by_xpath('//*[@id="reward-section"]/div/div[2]/div/div/div[4]/div/div[2]/p')
print(tag_fl.text)
all_data.insert(0,tag_fl.text)

print(all_data)
```
