```python
t_all_data = 'USD 1.636431'
aastockSPY_Jan = ''
percentage = r'\d*\.\d*'
aastockSPY_Jan = re.findall(percentage,t_all_data)
print(aastockSPY_Jan)
```

```python
#calclate the interest rate with the dividend
#interest rate of SPY in 90 ,180 ,360
#denominator is the price of the ETF
interest_SPY_Apr = round(float(aastockSPY_Apr[0])*100 / 411.867 , 2)
print(interest_SPY_Apr)
interest_SPY_Jul = round(float(aastockSPY_Jul[0])*100 / 434.211 , 2)
print(interest_SPY_Jul)
interest_SPY_Oct = round(float(aastockSPY_Oct[0])*100 / 456.219 , 2)
print(interest_SPY_Oct)
interest_SPY_Jan = round(float(aastockSPY_Jan[0])*100 / 448.515 , 2)
print(interest_SPY_Jan)
```

```python
ETF_SPY = [0,0,0,0,interest_SPY_Apr,round((interest_SPY_Apr+interest_SPY_Jul),2),round(interest_SPY_Apr+interest_SPY_Jul+interest_SPY_Oct+interest_SPY_Jan,2)]
print(ETF_SPY)
```
