```python
import pandas as pd
import numpy
```

```python
AAX = [6.35, 8.35, 10.35, 11.15, 12.15, 13.35, 14.5]
Binance = [0, 0, 5.0, 9.0, 9.0, 9.0, 9.0]
Huobi = [2.0, 5.5, 5.5, 6.0, 6.5, 6.5, 6.5]
Crypto_com = [2.0, 2.0, 4.0, 4.0, 8.0, 8.0, 8.0]
Bybit = [3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0]
HSBC = [0, 0, 0, 0, 0, 2.57, 3.89]
ETF_SPY = [0, 0, 0, 0, 0.31, 0.63, 1.3]
ETF_2800 = [0, 0, 0, 0, 0, 0.45, 2.75]
```

```python
df = pd.DataFrame()
df['AAX'] = AAX
df['Binance'] = Binance
df['Huobi'] = Huobi
df['Crypto_com'] = Crypto_com
df['Bybit'] = Bybit
df['HSBC'] = HSBC
df['ETF_SPY'] = ETF_SPY
df['ETF_2800'] = ETF_2800
df = df.T
print(df)
```

```python
df.to_csv("USDT_staking_return.csv")
```
